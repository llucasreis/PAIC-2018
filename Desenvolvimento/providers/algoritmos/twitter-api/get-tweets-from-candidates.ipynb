```python
initial_date = datetime.datetime(2018, 1, 1)
final_date = datetime.datetime(2018, 10, 28)
```

```python
def scrapy_tweets_from(screen_name):
    print(f'[INFO] Getting tweets from: {screen_name}')
    tweets = []
    
    all_tweets = api.user_timeline(screen_name=screen_name, count=200)
    
    while len(all_tweets) > 0:
        
        oldest = all_tweets[-1].id - 1
        
        print("[INFO] Processing tweets...")
        
        found = True
        
        for tweet in all_tweets:
            if tweet.created_at <= final_date and tweet.created_at >= initial_date:
                tweets.append(tweet)
            if found:
                print("[INFO] Found tweets in interval!")
                found = False
        
        print("[INFO] Getting older tweets...")
        
        all_tweets = api.user_timeline(screen_name=screen_name, count=200, max_id=oldest)
    
    print("[INFO] Proceeding to export to .csv")
    
    tweets_to_csv = [[tweet.text] for tweet in tweets]
    
    with open('%s_tweets.csv' % screen_name, 'w') as f:
        writer = csv.writer(f)
        writer.writerow(["text"])
        writer.writerows(tweets_to_csv)
    f.close()
    
    print("[INFO] Finished.")
```
